# 3. Ubah nama kolom:
a. CustomerID menjadi CustID
b. Semua Kolom menjadi lowercase (huruf kecil)

```python
#soal 3a, CustomerID menjadi CustID

df = df.rename(columns={'customerID':'CustID'})
df['CustID']
```

```python
#soal 3b, semua kolom menjadi lowercase
df.columns = df.columns.str.lower()
```

# 4. Hilangkan kata (automatic) dari value kolom PaymentMethod, pastikan juga tidak ada spasi yang tersisa di awal/akhir! (hint: gunakan string functions)

```python
df['paymentmethod'] = df['paymentmethod'].str.replace("automatic",'').str.replace("(","").str.replace(")","").str.strip()
df
```

# 5. Buatlah beberapa data preprocessing berikut:

```python
#untuk soal 5A
#mengambil tabel phone service
df_phoneservice = df[['phoneservice']]
#membuat one hot encoding dengan prefix service
pd.get_dummies(df_phoneservice, columns=['phoneservice'],prefix='service')
```

```python
#5B


#mengambil tabel monthly charge
df_monthlycharges = df[['monthlycharges']]

#Import MinMaxScaler

from sklearn.preprocessing import MinMaxScaler

#Normalization untuk kolom monthly charge

min_max_scaler = MinMaxScaler()

df['monthlychargesnormalize'] = min_max_scaler.fit_transform (df[['monthlycharges']])
df['monthlychargesnormalize'].describe()
```

```python
#5C

#ubah tipe data ke float

df['totalcharges'] = pd.to_numeric(df['totalcharges'],errors='coerce')


#Import StandartScaler

from sklearn.preprocessing import StandardScaler

# Standarization untuk kolom totalcharges

std_scaler = StandardScaler()

df['totalchargesSTD'] = std_scaler.fit_transform (df[['totalcharges']])
df['totalchargesSTD'].describe()
```
